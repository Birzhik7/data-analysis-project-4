```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
data=pd.read_csv('Cars Dataset.csv')
```

```python
data
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.dropna(inplace=True)
```

```python
data[data.duplicated(keep=False)]
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data.info()
```

```python
data['Price']=data['Price'].str.replace(',','',regex=False)
data['Price']=data['Price'].astype('int64')
```

```python
data['Year']=data['Year'].replace(20224.0,2024)
```

```python
data['Year']=pd.to_datetime(data['Year'],format='%Y')
data['Year']=data['Year'].dt.year
```

```python
data
```

```python
data.rename(columns={' Km/h ':'Km/h'},inplace=True)
```

```python
data['Km/h'].describe()
```

```python
data['high/low']=np.where(data['Price']>1000000,'High price','Low price')
```

```python
data['Model']=data['Model'].str.split()
```

```python
data['Model']=data['Model'].apply(' '.join)
```

```python
data
```

```python
def true_speed(speed):
    speed=speed.split(',')
    speed=speed[0].strip()
    return speed
```

```python
data['Km/h']=data['Km/h'].apply(true_speed)
data['Km/h']=data['Km/h'].astype('int64')
```

```python
data['Km/h']=data['Km/h'].where(data['Km/h']<400,other=250)
```

```python
data['Km/h']=data['Km/h'].where(data['Km/h']>100,other=110)
```
